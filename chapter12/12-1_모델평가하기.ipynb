# 모델 평가하기
- 우리의 목표는 학습하는 데이터셋에 대해 손실값을 최소화하는 것이덨다.
- 하지만 이 방향은 수집된 데이터에만 모델이 잘 동작하는 현상인 오버피팅에 빠질 수 있다.
- 따라서 이 장에서는 수집된 데이터셋을 잘 학습하면서도 오버피팅을 피할 수 있도록 하는 방법에 대해서 소개하도록 하겠다.

## 좋은 평가란?
- 모델을 실제 배포하고 나서 문제가 발생하는 것 보다 미리 잘 분비된 테스트 데이터셋을 통해 객관적인 평가를 수행하는 것이 중요하다.
- 테스트를 위한 데이터셋의 난이도 설정도 중요한 포인트중 하나다
    - 실제 서비스에서 모델이 에측할 데이터는 대부분 쉬운 수준의 샘플들일 것이다.
    - 그러다 가끔 어려운 데이터가 주어질 수 있다. 서비스의 품질은 여기에서 판가름 난다.
- 그렇기 때문에 테스트 데이터셋을 공들여 구성해야한다.
- 테스트 데이터셋은 적절한 수준의 샘플을 직접 엄선하여 구성하는 것이 바람직하다.
    - 너무 쉬우면 모든 곳에서 좋은 성능을, 너무 어려우면 모든 곳에서 나쁜 성능을 보여줄 것이니 난이도 조절이 중요하다
- 잘 못하면 노이즈가 섞이거나 너무 쉬운 테스트셋이 될 수 있다.
- 결국 이 평가를 통해서 타사의 서비스 또는 실험내의 베이스라인과 비교를 수행하게 된다


## 정성평가와 정량평가
- 정성평가
    - 정확하고 사람에 읳나 채점 방식으로 채점 기준이 모호한 경우에 필수다
- 정량평가 
    - 빠르고 정확하고 컴퓨터에 의한 자동 채점됨으로 정성평가와 비슷할수록 좋은 알고리즘이다.

## 실무에서 평가가 진행되는 과정의 예
- 학습이 완료된 모델에 테스트 데이터셋을 넣어 추론을 수행한다.
- 또한 동일한 테스트 데이터셋을 다른 베이스라인 모델에 넣어 추론을 수행한다. 
    - 이때 베이스라인 모델의 코드를 가지고 있을 수도 있지만 경쟁사 서비스와 같이 코드가 없는 경우에는 직접 서비스에 데이터를 넣어 결과를 얻을 수 있다.
- 채점 기준이 모호한 경우에는 정성평가를 수행한다.
    - 이때 중요한 것은 누가 어떤 예측 결과를 내놓았는지를 가리고 평가하는 블라인드 테스트를 수행해야 한다.
    - 만약 블라인드 테스트가 수행되지 않으면 이 테스트는 객관성을 갖추지 못했다고 볼 수 있다.
- 미리 잘 정의된 정성평가 기준에 따라 평가가 완료되면 결과를 취합하여 보고한다.

> 배포 주기에 따라 정성평가를 수행하고 정량평가는 수시로 수행될 것이다.
