# 오버피팅이란

## 목표 수정
- 좋은 인공지능이란 일반화를 잘하는 것이다.
- 보지 못한 데이터에 대해서 좋은 예측을 하는 것이 목표이다.
- 따라서 우리는 학습 오차를 최소화하는 것이 최종 목표가 아니다.
- 최종 목표는 일반화 오차를 최소화하는 것이다. 

## 오버피팅과 언더피팅
- 오퍼피팅
    - 학습 오차가 일반화 오차에 비해서 현격하게 낮아지는 현상
    - 학습데이터에 편향이나 노이즈가 있을 때 이것까지 모델이 학습함으로써 오히려 모델이 일반화를 수행하는 데 방해되는 상태를 뜻한다.
    - 오버피팅은 모델의 수용능력이 데이터의 복잡한 관계를 학습하기에 충분하다고 판단되기도 한다. 그래서 이후 오버피팅 방지하는 방향으로 모델을 개선하면 된다.
- 언더피팅
    - 모델이 충분히 데이터를 학습하지 못하여 학습 오차가 충분히 낮아지는 현상
    - 보통 모델의 수용 능력이 부족하여 복잡한 데이터의 관계를 학습하지 못하는 경우에 많이 발생한다. 
    - 혹은 모델의 수용력은 충분하짐나 적절한 최적화를 거치지 않아서 발생하기도 한다.

## 검증 데이터셋
- 가장 이상적은 것은 불필요한 특징까지 학습되는 순간이 올텐데 이것을 잘 감지하고 오버피팅으로 전환되는 순간에 학습을 멈추는 것이다. 
- 이것을 수행하기 위해서 우리는 검증 데이터셋을 도입한다. 
- 만약 모델이 오버피팅이 발생하지 않고 검증 데이터셋이 학습 데이터셋과 같은 성질(분포)를 갖고 있다면
- 모델은 검증 데이터셋에 대해서도 비슷한 손실 값을 가질 것이다. 
- 이룰 위해 학습데이터 일부를 임의로 나누어 검증 데이터셋으로 구성하고 이것을 학습하는 도중에 주기적으로 모델에 넣어 손실 값을 구함으로써 오버피팅 여부를 확인할 수 있다.
- 보통 매 에포크마다 이 작업을 수행하며 이 과정을 검증이라고 부른다. 

> 여기서 중요한 점은 검증 데이터셋을 통해서는 최적화(학습)를 진행하지 않으며, 데이터셋을 랜덤하게 나누는 과정에서 편향이 있으면 안된다.<br>
> 모델은 학습 데이터에대해서만 학습된 상태이므로 검증 데이터셋에 대해서 좋은 예측을 수행한다면 일반화가 잘되었다고 판단할 수 있다. 또는 검증 오차가 높담녀 일반화가 잘 되고 있지 않다고 판단할 수 있다.
