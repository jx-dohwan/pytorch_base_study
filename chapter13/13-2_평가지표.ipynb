# 평가지표
- 이산 데이터로 표현되기 때문에 손실 값 이외에도 정확도와 같은 다른 평가지표가 존재한다.

## 임곗값 설정과 그에 따른 상충 문제
- 문제의 성격에 따라 임곗값을 설정할 수 있다.
- 임곗값을 설정하기 위한 정책(policy)은 우리가 풀고자 하는 문제의 성격을 따른다.
- 참 클래스를 놓치지 않을지, 거짓 클래스를 놓치지 않을지 등에 따라 달라진다.

## 정확도 그리고 정밀도와 재현율
- 정확도
    - 가장 널리 쓰이는 지표이다.
    - 정확도는 전체 예측 개수와 전체 예측 중에서 실제로 맞춘 참의 갯수 비율을 나타낸다.
    - 정확도는 테스트 데이터셋의 클래스가 불균형할 경우 모델의 성능을 정확하게 반영하지 못한다.
    - 또한 클래스별 성능을 자세히 파악하고자 할 때 어려움을 겪을 수 있다.
- 정밀도
    - 모델이 양성 클래스라고 예측한 개수중에서 예측 중 실제로 맞춘 양성 클래스 개수의 비율
- 재현율
    - 실제로 양성 클래스인 샘플의 개수 중에서 예측 중 실제로 맞춘 양성 클래스 개수의 비율

> 같은 모델 안에서 임곗값을 바꾸게 되면 정밀도와 재현율이 서로 반대로 움직이는 것을 확인할 수 있다.

## F1 점수 
- 임곗값이 고정되어 있을 때 정밀도와 재현율을 기반으로 계산하게 된다.
- 만약 모델이 다양한 분야에 적용된다면 적용되는 애플리케이션에 따라 정책과 임곗값이 달라져 f1 점수도 바뀔 것이다.
- F1 점수가 항상 일정하게 유지되거나 여러 베이스라인들 중에서 순위가 유지된다면 상관없을 것이다.
- 그렇지 않담녀 사용자 입장에서는 F1 점수만으로 모델의 성능을 파악하기 어렵다고 봐야 할 것이다.

## AUBOC
- 변화하는 임곗값에 따른 모델의 성능이다. 
- 이것은 모델의 강인함을 알 수 있는 척도이다. 
- 두 분포 사이의 거리가 멀수록 모델이 임곗값 변화에 대해 강인하다고 볼 수 있다.
- 실제 서비스 베포를 고려한다면 다양한 관점에서 모델을 바라보고 평가해야한다.
- 이와 같이 다양한 지표를 활용하여 객관적인 모델의 성능을 평가할 수 잇다.
