# 특징이란?
- 어떻게 딥러닝 데이터를 학습하는지 이해하기 위해 특징의 개념에 대해 알아야 한다.
- 이제부터는 왜 학습이 잘되는지에 대해서 알아보도록 하겠다.

## 특징
- 모델이 원하는 출력 결과를 도출하는 입력을 다른 샘플들과 구분해낼 수 있는 능력이 필요하다.
- 그래서 데이터의 샘플을 잘 나타내는 특징을 추출하고 학습할 수 있어야 한다.
- 각 샘플별로 특징을 수치화해서 모델에게 제공한담녀 숫자를 분류하거나 샘플을 구분하는데 유용하게 활용될 것이다.


## 특징 추출 방법 : 머신러닝 vs 딥러닝
- 전통적인 방식의 머신러닝에서는 주로 사람이 직접 가정을 세우고 특징을 추출하는 방법을 설정한다.
    - 가설의 설정이 잘못되거나 생각지 못한 특징이 존재할 수 있다.
- 딥러닝에서는 데이터를 날 것으로 넣고 신경만 모델이 직접 특징을 파악하고 추출하는 과정을 거쳐 분류 또는 회귀 작업을 수행한다.
    - 원인 분석이 어렵다. 모델이 스스로 특징을 추출하기 때문에

## 특징 벡터
- 특징 벡터는 특징들을 모아서 벡터로 표현한 것을 말한다.
- 엑셀 파일과 같은 테이블로 구성된 데이터셋의 각행오 이에 해당될 수 있다.
    - 각 열은 벡터가 되면 차원을 이루고 벡터의 각 차원은 어떤 속성에 대한 수치를 나타낼 수 있다.
    - 이때 벡터의 각 속성에 대한 수치가 비슷할수록 비슷한 샘플이라고 생각할 수 있을 것이다.
    - 즉, 특징 벡터를 구성하면 이를 통해 샘플 사이의 거리(유사도)를 계산할 수 있다.
