# 오토 인코더
- 오토 인코더는 비선형 차원 축소를 수 있도록 하는 심층신경망이다.
- 오토인코더는 인코더와 디커도러 구성되어 있으며 
- 인코더를 통해 데이터의 차원 축소와 복원을 수행하며 이 과정에서 최대한 입력 샘플의 정보를 보존하도록 손실 압축을 진행한다.
- 디코더는 인코더의 중ㄱ나 결과물을 받아 입력 샘플과 같아지도록 압축 해제를 수행한다.
- 결과적으로 복원을 성공적으로 진행하기 위해서 오토인코더는 학습 데이터의 특징을 추출하는 방법을 자동으로 학습하게 된다.
- 참고로 오토인코더는 입력 샘플과 출력 벡터의 MSE 손실을 최소화하도록 학습한다.

## 인코더와 디코더의 역할
- 인코더는 고차원의 입력을 받아 복원에 필요한 정보를 중심으로 손실 압축을 진행한다.
- 그러므로 인코더가 뱉어낸 중간 결과물z는 입력 샘플보다 낮은 차원의 백터이다.
- 만약 정보량에 비해 차원이 충분히 크다면 손실되는 정보가 적어질 것이고 반대로 차원이 작다면 손실되는 정보가 많아질 것이다.
- 인코더와 디코더 사이의 병목 구간에 존재하는 인코더의 중간 결과물 z는 입력 샘플을 최대한 보존하고 있다.
- 따라서 디코더는 주어진 중간 결과물 벡터를 활용하여 입력 샘플을 최대한 복원할 수 있을 것이다.
- 이 과정에서 중간 결과물 벡터는 효울적인 압축을 수행해야 하므로 자연스럽게 고밀도 벡터로 표현된다.
- 또한 중간 결과물 벡터 또한 입력 샘플에 대한 특징 벡터라고 볼 수 있다.

## 은닉 표현
- 이렇게 얻어진 중간 결과물 벡터z를 입력 샘플x에 대한 은닉 표현이라고 부른다.
- 그리고 해당 벡터가 존재하는 공간을 은닉 공간이라고 부른다.
- 은닉 벡터는 원하는 출력을 뱉어내는 데 필요한 정볼르 해당 모델만 알고 있는 방법으로 감춰놓고 있다
- 중간 결과물 벡터도 당연히 입력 샘플에 대한 특징 벡터라고 볼 수 있으며 은닉 벡터라고도 할수 있다.
