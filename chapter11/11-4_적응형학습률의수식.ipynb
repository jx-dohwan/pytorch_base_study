# 적응형 학습률의 수식
- 손실 함수는 데이터셋의 샘플들에 대해 타깃 출력값과 함수의 출력값 사이의 차이를 곗나
- 이것을 가중치 파라미터에 대해서 미분하면 그래디언트 벡터를 얻을 수 있다.
- 앞서 구한 그래디언트에 학습률을 곱해 기존 파라미터에서 빼 준 결과를 다음 파라미터로 정한다.
- 여기에서 t는 파라미터 업데이트 횟수를 뜻한다. 즉 이터레션을 의미한다.
- 따라서 학습률은 그래디언트가 얼마나 파라미터 업데이트에 반영될지를 나타내며 
- 학습률의 크기에 따라 학습이 발산하거나 더디게 진행될 수 있다.
> 그러므로 데이터셋이 바뀌거나 모델 구조가 바뀌는 등의 상황에서는 매번 최적 학습률을 찾기 위한 튜닝 작업을 수행해야 한다.

## 모멘텀의 수식
- 이전까지의 모멘텀 그래디언트와 현재 그래디언트의 선형 결합, 즉 처음부터 현재까지의 디스카운트 파라미터가 곱해진 그래디언트의 누적 합
- 그리고 가중치 파라미터를 업데이트하기 위해 현재 가중치 파라미터에 모멘텀 그래디언트를 더해주게 된다.
> 모든 스텝(time-step)의 그래디언트를 누적하여 빼주는 형태<br>
> 여전히 수식에 학습률이 남아있다. 결국 누적된 그래디언트가 파라미터 업데이트에 끼치는 영향도를 설정하고 있음을 알 수 있다.

## 아다그래드의 수식
- 이제 자동을 학습률이 설정되도록 하겠다. 
- 학습 초반에는 큰 학습률을 통해 빠르게 학습되도록 하고
- 학습 후반에는 작은 학습률을 통해 파라미터를 미세하게 조정할 수 있도록 하고자 한다.
- 아다 그래드는 각 가중치 파라미터 별로 학습률이 정해지는 것을 기본으로 한다.
- 여기에 다음 수식과 같이 그래디언트의 제곱을 누적한 값을 계산하여 활용할 것이다.
- 이제 이 누적값을 초기 학습률에서 나눠줌으로써 동적으로 학습률이 설정될 수 있다.
- 이로 인해 그동안 업데이트의 크기가 큰 가중치 파라미터의 요소는 작은 학숩률이 곱해질 것이다.
> 이처럼 적응형 학습률을 사용하게 되면 학습 초기에는 큰 학습률과 학습 후기에는 작은 학습률을 가질 수 있게 된다.

## 아담의 수식
- 아담의 경우 기존 적응형 학습률 방식에 모멘텀이 추가된 것이 가장 큰 특징이다.
- 또한 기존 아다그래이드에서 볼 수 있던 학습 후반부의 학습 부진 현상을 해소하기 위한 기법들도 생겨나
- 기존 하이퍼파라미터인 학습률 이외에도 하이퍼파라미터가 2개나 더 추가되었다.
- 추가된 하이퍼파라미터드들은 기본 설정값을 지니고 있다, 당분간은 해당 HP들을 건드릴 필요는 없을 것이다.
> 아담 옵티마이저는 모멘텀과 적응형 학습률이 모두 포함되어 있어 수렴 속도도 굉장히 빠르고 학습률의 변화에도 강인한 모습을 보인다.
